```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

## Datasets

```python
rainfall = (pd.read_csv("rainfall-monthly-number-of-rain-days.csv"))
humidity = (pd.read_csv("relative-humidity-monthly-mean.csv"))
x = pd.merge(humidity,rainfall)
humid_rain = x.drop('month', axis='columns')
display(humid_rain)
```

```python
temperature = pd.read_csv("surface-air-temperature-monthly-mean.csv")
y = pd.merge(rainfall,temperature)
rain_temp = y.drop('month', axis='columns')
display(rain_temp)
```

```python
rain_temp["mean_temp"] = rain_temp["mean_temp"].astype(int)
rain_temp
```

## Overall Chart (1985-2021)

```python
humid_rain.plot.scatter(x= "no_of_rainy_days", y="mean_rh", color='red')
```

```python
sns.regplot(x="no_of_rainy_days", y= "mean_rh", data=humid_rain, ci=None)
```

**Explanation:** As the mean humidity increases,the number of rainy days increases. And as the no.of rainy days increases,the probability of a flash flood increases.

```python
humidity.plot.scatter(x= "month", y="mean_rh", color='green')
```

```python
rainfall.plot.scatter(x= "month", y="no_of_rainy_days", color='purple')
```

**Explanation:** When the no.of rainy days is low, the probability for vegetation fires is high and SCDF can be on standby for vegetation fires.

```python
y.plot.scatter(x="no_of_rainy_days", y="mean_temp",color="blue")
```

```python
sns.regplot(x="no_of_rainy_days", y="mean_temp",data=y, ci=None)
```

**Explanation:** As the number of rainy days increases,the mean temperature decreases.When the mean temperature is high, SCDF can be on standby to prepare for vegetation fires.

```python
temperature.plot.scatter(x="month",y="mean_temp",color='purple')
```
